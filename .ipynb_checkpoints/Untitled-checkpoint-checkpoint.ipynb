```python
def indicators(the_map):
    the_map.x_unit=4*450
    the_map.y_unit=4*450
    vois=op.neighbor_grid(the_map.data)
    moy=np.mean(vois,axis=(-1,-2))
    nmap=the_map.data-moy
    vois_big=op.neighbor_grid(nmap>5,5,bound_method='duplicate')
    surf_text=np.mean(vois_big,axis=(-1,-2))
    conv_map=np.mean(op.neighbor_grid(the_map.laplacien>0.05,5,bound_method='duplicate'),axis=(-1,-2))
    
    in_grad=skimage.measure.block_reduce(np.log(1+the_map.norme_grad),(50,50),np.mean)
    itexture=skimage.measure.block_reduce(surf_text,(50,50),np.mean)
    iconv=skimage.measure.block_reduce(conv_map,(50,50),np.mean)
    alt=skimage.measure.block_reduce(the_map.data,(50,50),np.mean)
    
    return [in_grad,itexture,iconv,alt]
```

```python
def run_on(f,tab):
    trslt=[]
    for a in tab:
        rs=[]
        for b in a:
            rs.append(f(b))
        trslt.append(rs)
    return trslt
def run_on_d1(f,tab):
    trslt=[]
    for b in tab:
        trslt.append(f(b))
    return trslt
def merge(tab):
    ah=np.hstack(tab[0])
    ab=np.hstack(tab[1])
    fin=np.vstack((ah,ab))
    return fin
def access(tab,k):
    m=len(tab[0])
    return [[tab[0][i][k] for i in range(m)],
                     [tab[1][i][k] for i in range(m)]]
def merges(tab):
    tabs=[]
    for k in range(len(tab[0][0])):
        tabs.append(merge(access(tab,k)))
    return tabs
        
```
