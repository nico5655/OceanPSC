```python
coords=[]
grilleCoords=np.zeros(masqueN.shape)
cpt=0
for x in range(masqueN.shape[1]):
    for y in range(masqueN.shape[0]):
        coords.append((x,y))
        grilleCoords[y,x]=cpt
        cpt+=1
```

```python
pointsNoirs=[]
pts=grilleCoords[masqueN==0]
for ind in pts.reshape(pts.size):
    pointsNoirs.append(coords[int(ind)])
pointsNoirs=np.array(pointsNoirs)
```

```python
XN=pointsNoirs.T[0]
YN=pointsNoirs.T[1]
```

```python
masque2N=masqueN.copy()
for x,y in coords:
    if masque2N[y,x] == 1:
        dist=np.sqrt(np.min((XN-x)**2+(YN-y)**2))
        masque2N[y,x]=dist
        continue
    masque2N[y,x]=0
```
